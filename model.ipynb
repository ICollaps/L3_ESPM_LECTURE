```python
from ultralytics import YOLO
import pandas as pd
import matplotlib.pyplot as plt
```

```python
model = YOLO("Best.pt")
```

```python
video_path = "video.mp4"
```

```python
results = model(video_path)
```

```python
data_list = []
for result in results:
    result = result.numpy()

    data = result.boxes.xywh[0]
    print(data)
    data_list.append(data)
```

```python
data_result = pd.DataFrame(data_list , columns=["x" , "y" , "largeur" , "hauteur"])
data_result
```

```python
plt.plot(data_result["x"] , data_result["y"])
```

```python
plt.plot(data_result["largeur"])
plt.plot(data_result["hauteur"])
```
